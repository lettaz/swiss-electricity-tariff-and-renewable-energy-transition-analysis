```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python

# target = 'Class'

### load some data with pandas

csv_url = 'data/energyreporter_historized/energyreporter_national_historized.csv'
dataset = pd.read_csv(csv_url)
```

```python
dataset
```

```python
features = ['solar_power_installed_kwp', 'electric_car_share']
dates_column_name = 'energyreporter_date'
```

```python
import matplotlib.dates as mdates

fig, ((ax, bx), (cx, dx)) = plt.subplots(2, 2)

ax.plot(dataset[dates_column_name], dataset[features[0]])
ax.set_ylabel(features[0])
ax.xaxis.set_major_locator(mdates.AutoDateLocator(minticks=5, maxticks=10))

bx.plot(dataset[dates_column_name], dataset[features[1]])
bx.set_ylabel(features[1])
bx.xaxis.set_major_locator(mdates.AutoDateLocator(minticks=5, maxticks=10))

cx.plot(dataset[dates_column_name], dataset[features[1]])
cx.set_ylabel(features[1])
cx.xaxis.set_major_locator(mdates.AutoDateLocator(minticks=5, maxticks=10))

plt.gcf().autofmt_xdate()  # Rotate and align the tick labels
plt.tight_layout()
plt.show()      
```
